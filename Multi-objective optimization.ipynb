The file `vns.py`contains the implementation of a generic Variable Neighborhood Size (VNS) algorithm for a multi-objective optimization problem. The file `knapsack.py`contains an instance of a bi-objective knapsack problem, where the two objectives are to maximize utility and minimize weight. Review the code before running this bnotebook, that runs the algorithm on the knapsack problem, and illustrates the solutions that have been generated.

```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import vns
from knapsack import knapsack, oneSack
```

```python
pickleFile = 'knapsackPareto.pickle'
```

```python
utility = np.array([80, 31, 48, 17, 27, 84, 34, 39, 46, 58, 23, 67])
weight = np.array([84, 27, 47, 22, 21, 96, 42, 46, 54, 53, 32, 78])
capacity = 500
```

```python
theKnapsack = knapsack(utility, weight, capacity)
emptySack = theKnapsack.emptySack()
```

Setting the parameter `archiveInputFile` below to `pickleFile` starts the algorithm with all solutions generated from a previous run. Here, we start from scratch for the sake of illustration. 

```python
thePareto = vns.vns(
    theKnapsack,
    [emptySack],
    maxNeighborhood=20,
    numberOfNeighbors=10,
    archiveInputFile=None,
    pickleOutputFile=pickleFile,
)
```

We now display the information contained in the Pareto set generated by the algorithm, and saved in the pickle file. 

```python
pareto = vns.paretoClass(_, archiveInputFile=pickleFile)
```

```python
print(f'Number of pareto solutions: {len(pareto.pareto)}')
```

```python
objectives = list(pareto.pareto)[0].objectivesNames
objectives
```

We first provide all the solutions in the Pareto set at the end of the algorithm.  

```python
counter = 0 
for p in pareto.pareto:
    counter += 1
    print(f'*************** Model {counter} ************************')
    print(f'Sack {p}, Weight: {p.objectives[0]}, Utility: {-p.objectives[1]}')
    print('\n')
```

The plot below illustrates all solutions considered by the algorithm. Each solution corresponds to one point in the graph. 

- The x-coordinate corresponds to the weight of the sack, and the y-coordinate to the negative utility of the sack. These are the two objectives that must be minimized. 
- The larger circles correspond to all solutions that are not dominated. They are in the Pareto set.
- The crosses corresponds to solutions that happened to be non dominated at some point during the course of the algorithm, but have been removed from the Pareto set afterwards, as a dominating solution has been identified.
- Finally, the small dots corresponds to solutions that have been considered, but rejected because dominated by another solution already in the set.

```python
par_obj = [p.objectives for p in pareto.pareto]
par_x, par_y = zip(*par_obj)
```

```python
con_obj = [p.objectives for p in pareto.considered]
con_x, con_y = zip(*con_obj)
```

```python
rem_obj = [p.objectives for p in pareto.removed]
rem_x, rem_y = zip(*rem_obj)
```

```python
x_buffer = 10
y_buffer = 0.1
```

```python
plt.axis([min(par_x)-x_buffer,
          max(par_x)+x_buffer,
          min(par_y)-y_buffer,
          max(par_y)+y_buffer])
plt.plot(par_x, par_y, 'o', label='Pareto')
plt.plot(rem_x, rem_y, 'x', label='Removed')
plt.plot(con_x, con_y, ',', label='Considered')
plt.xlabel(objectives[0])
plt.ylabel(objectives[1])
plt.legend()
plt.show()
```
