```python
import numpy as np
import matplotlib.pyplot as plt
import metropolishastings as mh
```

```python
def myprocess(state):
    next_state = np.random.randint(1, 4+1)
    return next_state, 0.25, 0.25
```

```python
def weight(state):
    w = {
        1: 20, 
        2: 8,
        3: 3,
        4: 1,
    }
    return w[state]
```

```python
the_mh = mh.MetropolisHastings(1, myprocess, weight)
```

```python
the_mh.generateDraws(1000)
```

```python
values, bins, plot = plt.hist(the_mh.sequence, bins=[1, 2, 3, 4, 5])
```
